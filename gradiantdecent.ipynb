#Gradiant Decent
This code is an example of gradiant decent algorithm, used on neural network's training.

It's based on the dotCSV video on youtube about it.

```
import numpy as np
import matplotlib.pyplot as plt
```

```
# This is the function we are going to optimize. Could be anyone
f = lambda vector: np.sin(0.5*vector[0]**2 - 0.25*vector[1]**2 +3)*np.cos(2*vector[0]+1 - np.e**vector[1])
res = 100

# x and y are the parameters, z is the output and p is a randon point
_X = np.linspace(-3,3,res) 
_Y = np.linspace(-3,3,res)
_Z = np.zeros((res,res))
rp = np.random.rand(2)*6 - 3

# fill z matrix
for ix,x in enumerate(_X):
  for iy,y in enumerate(_Y):
    _Z[iy,ix] = f([x,y])

#plot the graphic
plt.contourf(_X,_Y,_Z,50)
plt.plot(rp[0],rp[1],"o",c="red")
plt.colorbar()

#derivate the point to get the gradiant
h=0.001 
lr=0.01 #Learning rate
grad=np.zeros(2)

for _ in range(1000):
  for i,value in enumerate(rp):
    p=np.copy(rp)
    p[i]=p[i]+h
    deriv=(f(p)-f(rp))/h
    grad[i]=deriv
  rp=rp-lr*grad
  if (_%10==0):
    plt.plot(rp[0],rp[1],".",c="blue")

```

Red point is the start, and the way it tooks is shown on blue.

Notice that the algorithm can go to the deepest value of the function.
